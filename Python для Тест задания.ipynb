```python
df["date"] = pd.to_datetime(df["date"])

users_day1 = df[df["date"] == "2023-11-01"]["user_id"].unique()
users_day2 = df[df["date"] == "2023-11-02"]["user_id"].unique()

returned = set(users_day1) & set(users_day2)

retention = len(returned) / len(users_day1) * 100

round(retention,1)
```

```python
total_users = df["user_id"].nunique()

users_with_views = df[df["view_adverts"] > 0]["user_id"].nunique()

conversion = users_with_views / total_users * 100

round(conversion,1)
```

```python
import pandas as pd
from scipy import stats

df_ab = pd.read_csv("Данные для тестового задания - Данные АБ тестов.csv")

df_ab.head()
```

```python
df_ab["experiment_num"].unique()
```

```python
for exp in df_ab["experiment_num"].unique():
    
    data = df_ab[df_ab["experiment_num"] == exp]
    
    control = data[data["experiment_group"] == "control"]["revenue"]
    test = data[data["experiment_group"] == "test"]["revenue"]
    
    arpu_control = control.mean()
    arpu_test = test.mean()
    
    t_stat, p_value = stats.ttest_ind(control, test)
    
    print("Experiment:", exp)
    print("ARPU control:", round(arpu_control,2))
    print("ARPU test:", round(arpu_test,2))
    print("p-value:", p_value)
    print()
```
